```python
# importing Libraries
import numpy as np
import pandas as pd
```

```python
# reading data
df = pd.read_csv('/workspaces/Data_cleaning_project_1/DataSet/Cafe Sales.csv')
```

```python
# inspecting first few records
df.head()
```

```python
# inspecting lst few records 
df.tail()
```

```python
# data structure
df.info()
```

```python
# duplicate records 
print(f'Number of duplicate records: {sum(df.duplicated())}')
```

```python
# missing records in percentage per column
df.isna().sum() * 100 / df.shape [0]
```

```python
# unique records 
df_columns = set(df.columns)
exclude = {'Transaction ID'}
df_columns = df_columns.difference(exclude)

for column in df_columns:
    print(f'{column}:\n{df[column].unique()}', end='\n\n')
```

```python
#replacing errors and unknow values
df.replace(['ERROR', 'UNKNOWN'], np.nan, inplace=True)
```

```python
# missing records after replacing errors and unknown values with np.nan
df.isna().sum() * 100 / df.shape [0]
```

```python
# datatype conversion
df['Quantity'] = df['Quantity'].astype('Int64')
df[['Price Per Unit','Total Spent']] = df[['Price Per Unit','Total Spent']].astype('float')
df['Transaction Date'] = pd.to_datetime(df['Transaction Date'])
```

```python
df.info()
```

```python
# dropping records where Transaction Date is missing    
df = df.dropna(subset=['Transaction Date'])
df
```

```python
# replacing missing records in Payment Method and Location with INA
df[['Payment Method', 'Location']] = df[['Payment Method', 'Location']].fillna('INA')
```

```python
# replacing missing records in Item with mode value
df['Item'] = df['Item'].fillna(df['Item'].mode()[0])
```

```python
# replacing missing records with median in Quantity
df['Quantity'] = df['Quantity'].fillna(df['Quantity'].median())
```

```python
# inspecting Item for Price Per Unit 
df.groupby('Item')['Price Per Unit'].value_counts()
```

```python
# filling missing records in Price Per Unit with mode value of Item category
df['Price Per Unit'] = df.groupby('Item')['Price Per Unit'].transform(lambda x: x.fillna(x.mode()[0]))
```

```python
# filling missing records in Total Spent 
df['Total Spent'] = df['Total Spent'].fillna(df['Quantity'] * df['Price Per Unit'])
```

```python
df.info()
```

```python
# saving cleaned data to csv file
df.to_csv('/workspaces/Data_cleaning_project_1/DataSet/Cafe Sales Cleaned.csv', index=False)
```
